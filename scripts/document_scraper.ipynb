```python
import requests
import pandas as pd
import os
import time
import random
```

```python
start_date = "19290101"
end_date = "19391231"


date_range = pd.date_range(start=start_date, end=end_date, freq="D")

date_list = date_range.strftime("%Y%m%d").tolist()

print(date_list)
```

```python
url_list = []
for date in date_list :
    basic_url = "https://gallica.bnf.fr/ark:/12148/cb34363182v/date"
    date_url = basic_url+date
    url_list.append(date_url)
```

```python
for url in url_list :
    time.sleep(13)
    response = requests.get(url,allow_redirects=True)
    final_url = response.url
    response=requests.get(final_url+"{ark}.texteBrut")
    cherck_url=response.url
    text_save = response.text

    print(cherck_url)
    if cherck_url[46:50] == "date" :
        continue
    else :
        save_path = os.path.join(os.getcwd(), f"{url[-12:]}.html")
        with open(save_path, "w", encoding="utf-8") as file:
           file.write(text_save)
```
